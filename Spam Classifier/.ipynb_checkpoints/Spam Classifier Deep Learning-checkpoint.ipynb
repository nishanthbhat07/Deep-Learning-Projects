## Importing Packages

```python
from keras import Sequential
from keras.layers.core import Dense,Activation
from keras.optimizers import Adam
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical 
```

```python
plt.style.use('ggplot')
```

## Reading the data from the file

```python
data=pd.read_csv('spam.table.txt',sep='\t',names=['Label','SMS Messages'])
```

```python
data.head()
```

## Size of the dataset
## 5572 Rows and 2 columns

```python
data.shape
```

## Visualizing the number of ham and spam

```python
data.groupby('Label').count().plot(kind='bar')
```

```python
data.Label.replace({'ham':0,'spam':1},inplace=True)
```

```python
Y=to_categorical(data['Label'])
```

```python
data['SMS Messages']=data['SMS Messages'].str.strip().str.lower()
data['words']=data['SMS Messages'].str.split(" ")
```

```python
spam_data=data[data.Label==1]
```

```python
comment_words=''
for i in spam_data['words']:
    comment_words += " ".join(i)+" "
```

```python
stopwords = set(STOPWORDS) 
wordcloud = WordCloud(width = 800, height = 800, \
                background_color ='white', \
                stopwords = stopwords, \
                min_font_size = 10).generate(comment_words)
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```python
X_train,X_test,y_train,y_test=train_test_split(data['SMS Messages'],Y)
```

```python
vectorizer = CountVectorizer()
features_train_transformed = vectorizer.fit_transform(X_train.tolist())
features_test_transformed  = vectorizer.transform(X_test)
```

```python
(features_train_transformed.shape)
```

```python
features_train_transformed.todense().shape
```

```python
model=Sequential()
```

```python
model.add(Dense(32,input_dim=7475))
model.add(Activation('relu'))

model.add(Dense(64))
model.add(Activation('tanh'))

model.add(Dense(64))
model.add(Activation('tanh'))

model.add(Dense(128))
model.add(Activation('relu'))

model.add(Dense(2))
model.add(Activation('sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=0.0001),metrics=['accuracy'])
```

```python
history=model.fit(x=features_train_transformed.todense()[:3180],y=y_train[:3180],validation_data=(features_train_transformed.todense()[3180:],y_train[3180:]),epochs=10,batch_size=10)
```

```python
predictions=model.predict(features_test_transformed.todense())
rounded_prediction=np.argmax(predictions,axis=-1)
```

```python
cm=confusion_matrix(np.argmax(y_test,axis=-1),rounded_prediction)
```

```python
import itertools
def plot_confusion_matrix(cm,classes,normalize=False,title='Confusion Matrix',
                         cmap=plt.cm.Blues):
    plt.imshow(cm,interpolation='nearest',cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks=np.arange(len(classes))
    plt.xticks(tick_marks,classes,rotation=45)
    plt.yticks(tick_marks,classes)
    if normalize:
        cm=cm.astype('float')/cm.sum(axis=1)[:,np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix without normalization")
    print(cm)
    thresh=cm.max()/2
    for i,j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
        plt.text(j,i,cm[i,j],
                horizontalalignment='center',
                 color='white' if cm[i,j] > thresh else "black"
                )
        plt.tight_layout()
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
```

```python
cm_plots_labels=['ham','spam']
plot_confusion_matrix(cm=cm,classes=cm_plots_labels,title='Confusion Matrix')
```

```python
model.evaluate(features_test_transformed.todense(),y_test)
```
